```python
from sklearn.model_selection import GridSearchCV
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from scikeras.wrappers import KerasRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve
from warnings import simplefilter
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import logging
import time
import pickle
logging.basicConfig(filename='log/eMQTT-IA.log', encoding='utf-8', level=logging.DEBUG,format='%(asctime)s - %(levelname)s - %(message)s')
seed = 7
chunksize = 100000
```

```python
reader_train = pd.read_csv("FINAL_CSV/train70_augmented.csv", chunksize=chunksize, iterator=True)
dftrain_raw = pd.concat(reader_train, ignore_index=True)
simplefilter(action='ignore', category=FutureWarning)
```

```python
reader_test = pd.read_csv("FINAL_CSV/test30_reduced.csv", chunksize=chunksize, iterator=True)
dftest_raw = pd.concat(reader_test, ignore_index=True)
simplefilter(action='ignore', category=FutureWarning)
```

```python
filter = ['legitimate', 'dos']
dftrain = dftrain_raw[dftrain_raw['target'].isin(filter)]
dftest = dftest_raw[dftest_raw['target'].isin(filter)]

dftrain = dftrain.sort_values('target')
dftest = dftest.sort_values('target')

display(dftrain)
display(dftest)
```

```python
mapping_test = {}
mapping_train = {}
#train
class_names_train = dftrain.target.unique()
dftrain=dftrain.astype('category')
cat_columns_train = dftrain.select_dtypes(['category']).columns
dftrain[cat_columns_train] = dftrain[cat_columns_train].apply(lambda x: x.cat.codes)
x_columns = dftrain.columns.drop('target')
x_train = dftrain[x_columns].values
y_train = dftrain['target']
mapping_train = dict(zip(dftrain.target.unique(), class_names_train))
logging.debug('Mapping train: ',mapping_train)
dftrain = dftrain.sample(frac=1, random_state=42)


#test
class_names_test = dftest.target.unique()
dftest=dftest.astype('category')
cat_columns_test = dftest.select_dtypes(['category']).columns
dftest[cat_columns_test] = dftest[cat_columns_test].apply(lambda x: x.cat.codes)
x_columns = dftest.columns.drop('target')
x_test = dftest[x_columns].values
y_test = dftest['target']
mapping_test = dict(zip(dftest.target.unique(), class_names_test))
logging.debug('Mapping test: ',mapping_test)
logging.debug("Ready to generate train and test datasets")
dftest = dftest.sample(frac=1, random_state=42)
```

```python
start = time.time()

#Decision tree
logging.debug("Starting Decision tree")
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
end = time.time()
diff=end-start
logging.debug("Training time: " + str(diff))
start = time.time()
starttest = time.time()
y_pred_dt = clf.predict(x_test)
y_pred_dt_roc = clf.predict_proba(x_test)
endtest =time.time()
difftest = endtest-starttest
logging.debug("Test time: " + str(difftest))

#Random Forest
logging.debug("Starting Random Forest")
clfrf = RandomForestClassifier()
clfrf = clfrf.fit(x_train, y_train)
end = time.time()
diff=end-start
logging.debug("Training time: " + str(diff))
start = time.time()
starttest = time.time()
y_pred_rf = clfrf.predict(x_test)
y_pred_rf_roc = clfrf.predict_proba(x_test)
endtest =time.time()
difftest = endtest-starttest
logging.debug("Test time: " + str(difftest))

#Multi layer perceptron
logging.debug("Starting Multi layer perceptron")
modelmlp = MLPClassifier( max_iter=130, batch_size=1000, alpha=1e-4, activation = 'relu',solver='adam', verbose=10, tol=1e-4, random_state=seed)
modelmlp.fit(x_train, y_train)
end = time.time()
diff=end-start
logging.debug("Training time: " + str(diff))
starttest = time.time()
y_pred_mlp = modelmlp.predict(x_test)
endtest =time.time()
difftest = endtest-starttest
logging.debug("Test time: " + str(difftest))

#Gradient boost
logging.debug("Starting Gradient boost")
modelgb = GradientBoostingClassifier(n_estimators=20, random_state=seed,verbose=2)
modelgb.fit(x_train, y_train)
end = time.time()
diff=end-start
logging.debug("Training time: " + str(diff))
starttest = time.time()
y_pred_gradient = modelgb.predict(x_test)
endtest =time.time()
difftest = endtest-starttest
logging.debug("Test time: " + str(difftest))

#Create Naive Bayes Classifier
logging.debug("Starting Naive Bayes")
gnb = GaussianNB()
gnb.fit(x_train, y_train)
end = time.time()
diff=end-start
logging.debug("Training time: " + str(diff))
starttest = time.time()
y_pred_nb = gnb.predict(x_test)
endtest =time.time()
difftest = endtest-starttest
logging.debug("Test time: " + str(difftest))

#Create KNN Classifier
logging.debug("Starting KNN")
clf_knn = KNeighborsClassifier()
clf_knn.fit(x_train, y_train)
end = time.time()
diff=end-start
logging.debug("Training time: " + str(diff))
starttest = time.time()
y_pred_knn = clf_knn.predict(x_test)
endtest =time.time()
difftest = endtest-starttest
logging.debug("Test time: " + str(difftest))

##Neural network
#logging.debug("Starting Neural network")
#model = Sequential()
#model.add(Dense(50, input_dim=x_train.shape[1], kernel_initializer='normal', activation='relu'))
#model.add(Dense(30, input_dim=x_train.shape[1], kernel_initializer='normal', activation='relu'))
#model.add(Dense(20, kernel_initializer='normal'))
#model.add(Dense(6,activation='softmax'))
#model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
#monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=5, verbose=1, mode='auto')
#history = model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[monitor],verbose=2,epochs=200,batch_size=1000) 
#end = time.time()
#diff=end-start
#logging.debug("Training time: " + str(diff))
#starttest = time.time()
#y_pred_nn = model.predict(x_test)
#y_pred_nn = np.argmax(y_pred_nn,axis=1)
#endtest =time.time()
#difftest = endtest-starttest
#logging.debug("Test time: " + str(difftest))
```

```python
logging.debug("Decision Tree, accuracy: " + str(metrics.accuracy_score(y_test, y_pred_dt)) + " F1 score:" + str(metrics.f1_score(y_test, y_pred_dt,average='weighted')))
matrixdt = confusion_matrix(y_test,y_pred_dt)
logging.debug(matrixdt)

logging.debug("Random Forest, accuracy: " + str(metrics.accuracy_score(y_test, y_pred_rf)) + " F1 score:" + str(metrics.f1_score(y_test, y_pred_rf,average='weighted')))
matrixrf = confusion_matrix(y_test,y_pred_rf)
logging.debug(matrixrf)

logging.debug("MultiLayerPerceptron, accuracy: " + str(metrics.accuracy_score(y_test, y_pred_mlp)) + " F1 score:" + str(metrics.f1_score(y_test, y_pred_mlp,average='weighted')))
matrixml = confusion_matrix(y_test,y_pred_mlp)
logging.debug(matrixml)

logging.debug("Naive Bayes, accuracy: " + str(metrics.accuracy_score(y_test, y_pred_nb)) + " F1 score:" + str(metrics.f1_score(y_test, y_pred_nb,average='weighted')))
matrixnv = confusion_matrix(y_test,y_pred_nb)
logging.debug(matrixnv)

logging.debug("GradienBoost, accuracy: " + str(metrics.accuracy_score(y_test, y_pred_gradient)) + " F1 score:" + str(metrics.f1_score(y_test, y_pred_gradient,average='weighted')))
matrixgb = confusion_matrix(y_test,y_pred_gradient)
logging.debug(matrixgb)

logging.debug("Starting KNN, accuracy: " + str(metrics.accuracy_score(y_test, y_pred_knn)) + " F1 score:" + str(metrics.f1_score(y_test, y_pred_knn,average='weighted')))
matrixknn = confusion_matrix(y_test,y_pred_knn)
logging.debug(matrixgb)
```
